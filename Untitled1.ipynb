```python

import os
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Conv2D ,Dense, Dropout, MaxPooling2D,Flatten,BatchNormalization,GlobalAveragePooling2D
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
base_dir = "/content/dermnet"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
```

```python
ata_test=keras.utils.image_dataset_from_directory(
    directory='/content/dermnet/test',
    labels='inferred',
    label_mode='categorical',
    batch_size=32,
    image_size=(224,224)
)
data_train=keras.utils.image_dataset_from_directory(
    directory='/content/dermnet/train',
    labels='inferred',
    label_mode='categorical',
    batch_size=32,
    image_size=(224,224)
)
```

```python
def process(image,label):
  image=tf.cast(image/255. ,tf.float32)
  return image,label

train=data_train.map(process)
test=data_test.map(process)
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
```

```python
base_model = ResNet50(
    weights='imagenet',
     include_top=True, # Remove original classifier
    input_shape=(224, 224, 3)
)
for layer in base_model.layers[-4:]:  # Unfreeze the last 4 layers
    layer.trainable = True
#base_model.trainable = True
```

```python
model = Sequential()
model.add(base_model)

#model.add(GlobalAveragePooling2D())
model.add(Flatten())
#model.add(Dense(2048, activation='relu'))
model.add(Dense(1024, activation='relu'))

model.add(BatchNormalization())


#model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(23, activation='softmax'))
```

```python
# Compile the model
model.compile(
    optimizer=Adam(learning_rate=1e-4),  # Lower LR because base model is frozen
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
model.fit(train, validation_data=test, epochs=30)
```

```python
model.save('skin_disease_model.h5')
```
